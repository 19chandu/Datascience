```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.countplot(x='sex',data=tips)
sns.set_style('darkgrid')
```

```python
sns.countplot(x='sex',data=tips)
sns.set_style('ticks')
sns.despine(left=True,bottom=True)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,3))
sns.countplot(x='sex',data=tips)
```

```python
sns.set_context('poster',font_scale=1)#notebook
sns.countplot(x='sex',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='seismic')#matplotlib colorplot
```
