```python
import seaborn as sns
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris
```

```python
sns.pairplot(iris)
```

```python
g=sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.kdeplot)
```
