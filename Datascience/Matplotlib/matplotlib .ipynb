```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np 
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
#functional
plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'g')
```

```python
# oo
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('LARGER PLOT')

axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')

```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=3,ncols=3)

plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')
```

```python
#FIGURE SIZE AND DIP(DOTS PER INCH)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))
plt.tight_layout()
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
fig.savefig('my_picture.png',dpi=200)
```

```python
fig = plt.figure()

ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Squared')
ax.set_title('TITLE')
ax.plot(x,x**3,label='X cubed')
ax.legend(loc=1)
```

```python
#Plot Appearance
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='green',linewidth=3,linestyle='--') # RGB hex code
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',linewidth=3,linestyle='-',marker='o',markersize='20',
       markerfacecolor='orange',markeredgewidth=2,markeredgecolor='blue') 
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='orange',linewidth=3,linestyle='--')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
